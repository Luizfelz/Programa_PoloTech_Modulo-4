### Questão 1

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Calcule a frequência absoluta para os sobreviventes no _Titanic_.

__Dica.:__ Utilize a função `.value_counts()`

Carregando as principais bibliotecas que iremos utilizar:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
df_freq_absol = df[df['survived'] == 1].value_counts()
df_freq_absol
```

### Questão 2

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Calcule a frequência relativa, relativa percentual e acumulativa para os sobreviventes no _Titanic_.

__Dica.:__ Utilize a função `.value_counts()`

```python
# frequência relativa:
df_freq_relat = df_freq_absol / len(df_freq_absol)
df_freq_relat
```

```python
# frequência relativa percentual:
df_freq_relat_per = df_freq_absol / len(df_freq_absol) * 100
df_freq_relat_per
```

```python
# frequência acumulativa
df_freq_cumul = (df[df['survived'] == 1].value_counts()).cumsum()
df_freq_cumul
```

### Questão 3

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Utilizando a coluna `Age` do _dataset Titanic_, defina as principais métricas estatísticas para essa variável tais como:
- Média;
- Desvio Padrão;
- Mínimo;
- Primeiro Quartil;
- Segundo Quartil (Mediana);
- Terceiro Quartil;
- Distância Interquartil (IQR);
- Máximo;
- Skewness;
- Moda.

```python
# a média, o desvio, valor mínimo, valor máximo e os quartis podem todos ser vistos com o uso do .describe(), como abaixo
df['age'].describe()
```

```python
# distância interquartil (IQR)
df_ordenado = df['age'].sort_values()
df_ordenado.dropna(inplace=True)
quartil_25 = np.quantile(df_ordenado, 0.25)
quartil_75 = np.quantile(df_ordenado, 0.75)
iqr = quartil_75 - quartil_25
print(f'IQR = {iqr}.')
```

```python
# skewness
df_ordenado.skew()
```

```python
# moda
df_ordenado.mode() # ou ainda: df_ordenado.mode()[0]
```
