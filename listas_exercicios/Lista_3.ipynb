### Questão 2

Neste exercício vamos demonstrar a importância de uma amostragem estratificada quando temos grupos significantes dentro da população. Faça os seguintes itens:

__A)__ Crie 3 listas de números aleatórios sendo elas:
    - uma lista com números aleatórios entre 40 e 50 contendo 10.000 elementos;
    - uma lista com números aleatórios entre 10 e 20 contendo 6.000 elementos;
    - uma lista com números aleatórios entre 80 e 90 contendo 4.000 elementos.


```python
import numpy as np
a = np.random.randint(40,51,10000)
b = np.random.randint(10,21,6000)
c = np.random.randint(80,90,4000)
```

    
__B)__ Calcule a média e a variância para cada uma das 3 listas;


```python
# média da lista a
media_a = a.mean()
variancia_a = np.var(a)
print(f'Média "a"= {media_a}')
print(f'Variância "a"= {variancia_a} \n')

# média da lista b
media_b = b.mean()
variancia_b = np.var(b)
print(f'Média de "b"= {media_b}')
print(f'Variância "b"= {variancia_b} \n')

# média da lista c
media_c = c.mean()
variancia_c = np.var(c)
print(f'Média "c"= {media_c}')
print(f'Variância "c"= {variancia_c}')
```


__C)__ Calcule a média e a variância para a população, onde a população seja as 3 listas juntas;

__Dica.:__ Neste caso, temos que trabalhar com média e variância ponderada!

Para calcular a média ponderada devemos seguir a fórmula abaixo:
    
$$\overline{x}_{p} = \sum_{i=1}^{k} p_i \overline{x}_i$$

onde $k$ e o número de grupos dentro da população e $p_i$ os pesos de cada grupo definido por $p_i = n_i/N$

Já para o variância ponderada, deve-se calcular da seguinte forma:

$$S_{p}^{2} = \sum_{i=1}^{k} S_{i}^{2}$$

```python
# tamanho total da população:
N = len(a) + len(b) + len(c)

# pesos de cada uma das populações:
p_a = len(a)/N
p_b = len(b)/N
p_c = len(c)/N

# média da população:
media_pop = (p_a * media_a) + (p_b * media_b) + (p_c * media_c)

# variância da população (somatório das variâncias amostrais):
var_pop = variancia_a + variancia_b + variancia_c

print(f'A média da população é = {round(media_pop,4)} e a variância da população é = {round(var_pop,4)}.')
```


__D)__ Una as 3 listas em uma só e retire uma amostra de 1% da lista resultante. Calcule a média e a variância para esta amostra;


```python
import random
# população total = soma das três amostras
populacao = list(a) + list(b) + list(c)

# tamanho da amostra equivalente a 1% (=0.01):
tam_amostra = int(0.01 * len(populacao))

# amostra aleatoria, de tamanho igual a 'tam_amostra', da população total:
amostra = np.array(random.sample(populacao,tam_amostra))

# média e variância:
media = amostra.mean()
var = np.var(amostra)

print(f'A média da população é = {round(media,4)} e a variância é = {round(var,4)}.')
```


__E)__ Por fim, faça uma amostra estratificada da seguinte forma: faça uma amostra aleatória de 1% de cada uma das 3 listas e em seguida calcule a média e a variância para a amostra estratificada (será necessário utilizar a média e variância ponderada!).


```python
# retirando amostras de 1% de cada uma das amostras iniciais:
amostra_1 = np.array(random.sample(list(a),int(0.01*len(a))))
amostra_2 = np.array(random.sample(list(b),int(0.01*len(b))))
amostra_3 = np.array(random.sample(list(c),int(0.01*len(c))))

# média e variância de cada amostra:
media_1 = amostra_1.mean()
var_1 = np.var(amostra_1)

media_2 = amostra_2.mean()
var_2 = np.var(amostra_2)

media_3 = amostra_3.mean()
var_3 = np.var(amostra_3)

# média e variância ponderada:
media = (p_a * media_1) + (p_b * media_2) + (p_c * media_3)
var = var_1 + var_2 + var_3

print(f'A média ponderada é = {round(media,4)} e a variância é = {round(var, 4)}.')
```


__F)__ Compare o resultado das métricas para a amostra e a amostra estratificada com o valor das métricas da população.

```python
'''
Na alternativa D, fizemos uma amostragem simples, enquanto que na alternativa E fizemos uma amostragem estratificada. Comparando o resultado dessas duas alternativas com o 
resultado adquirido na alternativa C, vemos que com a amostragem estratificada obtivemos valores bem mais próximos do que com a amostragem simples.
'''
```

### Questão 3

 Uma máquina de sorvete está regulada de modo a servir uma média de 120g por casquinha. Se a quantidade servida por casquinha seguir uma distribuição normal com desvio padrão de 18g, determine a porcentagem de casquinhas que conterão mais de 150g de sorvete.

```python
import numpy as np
import scipy.stats as st
# qual a probabilidade de ter x > 150, ou seja P(x > 150)?
# P(x > 150) = 1 - P(x <= 150)
# onde P(x > 150) = P(z > (x - media) / desvpad)
# logo: P(x <= 150) = P(z > (x - media) / desvpad)
# Portanto:

media = 120
x = 150
desvpad = 18
z = (x - media) / desvpad
prob_z = st.norm.cdf(z)

print(f'A probabilidade de conter mais de 150g é de {round(1 - prob_z,4) * 100}%.')
```

### Questão 4

O peso médio de 500 estudantes do sexo masculino de uma determinada universidade é 71 Kg e o desvio padrão é 5,3 Kg. Admitindo que os pesos são normalmente distribuídos, determine a oercentagem de estudantes que pesam entre 65 e 72,5 kg

```python
# P(65kg <= x <= 72,5kg) = P(x <= 72,5kg) - P(x <= 65kg)
media = 71
desvpad = 5.3

# Primeiro, P(x <= 72,5kg):
x_1 = 72.5
z = (x_1 - media) / desvpad
prob_z1 = st.norm.cdf(z)

# Segundo, P(x <= 65kg):
x_2 = 65
z = (x_2 - media) / desvpad
prob_z2 = st.norm.cdf(z)

print(f'A probabilidade de estar entre 65 e 72,5 kg é de {round((prob_z1 - prob_z2) * 100,4)}%.')
```
