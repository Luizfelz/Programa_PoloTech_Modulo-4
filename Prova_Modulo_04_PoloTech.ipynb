**Questão 4:** Em uma filial de um banco, 40% dos clientes que vão ao caixa eletrônico sacam dinheiro do caixa. Utilizando de uma distribuição binomial, qual é a probabilidade de no máximo 12 clientes sacarem dinheiro em um dia que tivemos 30 clientes no caixa eletrônico?

[X] 57,8 %

[_] 52,2 %

[_] 42,2 %

[_] 47,8 %

[_] 40,0 %

```python
# Resolução:
from scipy import stats
p = 0.4   # 40%
n = 30    # quantidade total de clientes
k = 12    # valor procurado (quantidade de clientes que retirariam dinheiro do caixa)

rv = stats.binom(n, p)
resposta = rv.cdf(k) * 100

print(f'Probabilidade procurada = {round(resposta, 2)}%')
```

OBS: foi utilizado o CDF e não o PMF, pois, o valor que gostaríamos de encontrar é o de NO MÁXIMO 12 clientes retirem dinheiro do caixa eletrônico, de um total de 30 clientes que foram ao caixa eletrônico. Logo, gostaríamos do cumulativo das probabilidade de: apenas 1 cliente retirar dinheiro, ou 2 clientes, ou 3, e assim sucessivamente. Portanto, gostaríamos do valor acumulado até NO MÁXIMO 12 clientes. Logo, utilizamos o CDF (Cummulative Distribution Function).

**Questão 5:** Em um centro de distribuição logístico da Amazon, chegam cerca de 1.440 produtos por dia. Sabendo que segue uma distribuição de Poisson, qual a probabilidade de que em um minuto cheguem 3 produtos?


[_] 36,8 %

[_] 55,2 %

[_] 24,5 %

[X] 6,1 %

[_] 18,4 %

```python
# Resolução:
from scipy import stats

h = 24 * 60       # quantidade de minutos em 1 dia: 24 horas * 60 minutos
lbd = 1440 / h    # quantidade (taxa) de produtos que chegam em 1 minuto
k = 3             # frequência desejada

rv = stats.poisson(lbd)
resposta = rv.pmf(k) * 100

print(f'Probabilidade procurada = {round(resposta, 2)}%')
```

OBS: foi utilizado o PMF e não o CDF, pois, o valor que gostaríamos de encontrar é EXATAMENTE o de apenas 3 produtos chegarem em 1 minuto. Logo, gostaríamos do valor exato, e não o valor de probabilidade de até 3 produtos chegarem. Assim, utilizamos o PMF (Probability Mass Function).

image.png

**Questão 7:** Uma máquina de sorvete está regulada de modo a servir uma média de 80g por casquinha. Se a quantidade servida por casquinha seguir uma distribuição normal com desvio padrão de 8g, qual a porcentagem de casquinhas que conterão mais de 90g de sorvete?


[_] 26,6 %

[_] 73,4 %

[X] 10,6 %

[_] 50,0 %

[_] 89,4 %

```python
# Resolução:
from scipy import stats

mu = 80     # média populacional
sigma = 8   # desvio padrão populacional
x = 90      # valor esperado

# Seguindo a fórmula para a distribuição normal em que: z = (x - mu) / sigma, temos:
z = (x - mu) / sigma

# Logo:
probabilidade = (1 - stats.norm.cdf(z)) * 100

print(f'A probabilidade de casquinhas com mais de 90g é = {round(probabilidade, 2)}%.')
```

OBS: A probabilidade foi calculada como 1 menos a probabilidade de Z, pois estamos procurando o valor que representa o quantitativo acima do valor de Z, onde Z representa todas as probabilidades de casquinhas com ATÉ 90g (ou seja, < 90g).

**Questão 8:** O tempo para fabricar um perfume em uma empresa é descrito por uma variável aleatória X, medida em horas, com distribuição normal de média mu = 7 e desvio padrão sigma = 2. Qual a probabilidade de que um perfume será finalizado entre 4 e 11 horas?


[X] 91,0 %

[_] 93,4 %

[_] 9,0 %

[_] 97,7 %

[_] 2,3 %

```python
# Resolução:
from scipy import stats

mu = 7
sigma = 2
x1 = 4    # valor esperado mínimo
x2 = 11   # valor esperado máximo

# Seguindo a fórmula para a distribuição normal em que: z = (x - mu) / sigma, temos:
z1 = (x1 - mu) / sigma
z2 = (x2 - mu) / sigma

# Logo:
probabilidade = (stats.norm.cdf(z2) - stats.norm.cdf(z1)) * 100

print(f'A probabilidade de ser finalizado entre 4 e 11 horas é = {round(probabilidade, 2)}%.')
```

OBS: Como queremos o valor entre 4 e 11 horas, devemos pegar a máxima probabilidade (z2) e subtrair a mínima probabilidade (z1), para obtermos o intervalo no qual teremos a probabilidade desejada.

**Questão 10:** Um fabricante de refrigerante anunciou que o índice de cafeína do seu refrigerante de cola não é maior do que 35mg por lata. Um laboratório realizou 20 análises do índice de cafeína obtendo os resultados (37, 34, 36, 37, 35, 32, 30, 35, 34, 36, 39, 38, 34, 35, 31, 30, 33, 35, 36, 31). Sabe-se que o índice de cafeína do refrigerante de cola dessa marca se distribui normalmente com variância igual a 4. Qual é, respectivamente, a conclusão do teste e o valor crítico para este teste com nível de significância de 5%? Utilize a suposição de que queremos identificar se a média de cafeína, caso não seja 35mg, seja maior que esse valor (caso unicaudal do teste). Obs.: não utilize a função ztest ou t.

[_] Não Rejeita H0, X_critico = 34,40 mg

[_] Rejeita H0, X_critico = 34,40 mg

[X] Não Rejeita H0, X_critico = 35,74 mg

[_] Rejeita H0, X_critico = 35,74 mg

```python
# Resultados:
from scipy import stats
import numpy as np

mu = 35       # média populacional
sigma = 2     # sigma = sqrt(variância) = sqrt(4) = 2
alpha = 0.05  # significância de 5%
amostra = [37, 34, 36, 37, 35, 32, 30, 35, 34, 36, 39, 38, 34, 35, 31, 30, 33, 35, 36, 31]
n = 20
x_obs = np.mean(amostra)
conf = 1 - alpha  # confiança

# Consideramos que H0: mu = 35 e H1: mu > 35. Portanto, primeiramente calculamos o valor do z:
z = (conf)
prob = stats.norm.cdf(z)

#Depois, pela fórmula, vamos calcular o valor do x_crítico:
x_crit = (z * (sigma / np.sqrt(n))) + mu

print(f'O valor do x_crit é = {round(x_crit, 2)}.')
print(f'O valor do x_obs é = {x_obs}.')

if (x_obs < x_crit):
  print(f'O valor de x_obs é menor que x_crit, portanto ACEITAMOS H0.')
else:
  print(f'O valor de x_obs é maior que x_crit, portanto REJEITAMOS H0.')
```
